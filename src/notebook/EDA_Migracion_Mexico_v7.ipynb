### Unificación de valores de las tablas <a id="unificacion"></a>

*En esta sección hacemos la combinación de las tablas 07 y 5. Sin embargo, podemos ver que se han creados una **segunda limpieza** tendrá que llevarse acabo*

> Antes de continuar con la limpieza se ha observado que las filas han incrementadp de 1683 en cada tabla a 85833 por lo que procedemos a verificar
la solución que se encontró fue crear una nueva columna llamada "Unicode" la cual ayuda a unificar mediante un único valor que tenga como referencia

>> Nos percatamos que los datos se están combinando de manera equivocada, por lo que procedemos al análisis por lo que se ha pensado en alguna forma de poder asegurarnos que cuando se hace el merge de ambas tablas, los datos se combinen de la misma manera puesto que anteriormente cuando se combinaba una tabla se pegaba con los grupos quinquenales ordenado por grupor y la otra tabla se combinaba pero ordenada por Género. ### Inicio del Análsis  <a id="analisis"></a>

Verificaremos si los motivos que conllevan a la migración de México influyen en que la migración sea en los Estados Unidos de América u en otros países.




```python
# Primero dividir las variables por tipo, usando la función que se programo en clase en los ejercicios
def card_tipo(df,umbral_categoria = 10, umbral_continua = 30):
    # Primera parte: Se prepara el dataset con cardinalidades, % variación cardinalidad, y tipos
    df_temp = pd.DataFrame([df.nunique(), df.nunique()/len(df) * 100, df.dtypes]) # Cardinaliad y porcentaje de variación de cardinalidad
    df_temp = df_temp.T # Como da los valores de las columnas en columnas, y quieremos que estas sean filas, las trasponemos
    df_temp = df_temp.rename(columns = {0: "Card", 1: "%_Card", 2: "Tipo"}) # Cambiamos el nombre de la transposición anterior para que tengan más sentido, y usamos asignación en vez de inplace = True (esto es arbitrario para el tamaño de este dataset)

    # Corrección para cuando solo tenemos un valor
    df_temp.loc[df_temp.Card == 1, "%_Card"] = 0.00

    # Creamos la columna de sugerenica de tipo de variable, empiezamos considerando todas categóricas pero podría haber empezado por cualquiera, siempre que adaptamos los filtros siguientes de forma correspondiente
    df_temp["tipo_sugerido"] = "Categorica"
    df_temp.loc[df_temp["Card"] == 2, "tipo_sugerido"] = "Binaria"
    df_temp.loc[df_temp["Card"] >= umbral_categoria, "tipo_sugerido"] = "Numerica discreta"
    df_temp.loc[df_temp["%_Card"] >= umbral_continua, "tipo_sugerido"] = "Numerica continua"

    return df_temp
```
